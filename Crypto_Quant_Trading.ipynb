````python
# Format the reading of all files in raw_csv folder
path = 'raw_csv/'
extension = 'csv'
os.chdir(path)
files = glob.glob('*.{}'.format(extension))
````

````python
# Read all csv files as a DataFrame and add them to file_dict
file_dict = {}

# Loop to read each CSV, format as DataFrame, and add to file_dict
for file in files:
  key = file
  df = pd.read_csv(file, header=1, infer_datetime_format=True, parse_dates=True).set_index('date') 
  
  file_dict[key] = df

    
#Old Attempts at formatting the Dataframe from CSV 
    #df = pd.read_csv(file, index=date, header=1).set_index('date')
    #usecols = ['date', 'open', 'high', 'low', 'close' 'Volume USD']
    
    # Set time column to date time
    # df['date'] = pd.to_datetime(df['date'])
    
    # Set time column to index
    # df = df.set_index('date')
````
